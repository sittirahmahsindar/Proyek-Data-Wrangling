Dataset **Diabetes (Pima Indians Diabetes Database)** terdiri dari **768 baris** dan **9 kolom**. Setiap baris merepresentasikan satu individu atau pasien yang menjadi objek penelitian. Dari total 9 kolom tersebut, delapan kolom pertama merupakan variabel independen (fitur), yaitu Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, dan Age, sedangkan satu kolom terakhir yaitu Outcome merupakan variabel dependen (target) yang menunjukkan hasil diagnosis diabetes, dengan nilai 0 untuk tidak diabetes dan 1 untuk diabetes.

Berdasarkan **informasi tipe data**, seluruh variabel dalam dataset ini bertipe numerik, yang terdiri dari 7 kolom bertipe **int64** dan 2 kolom bertipe **float64**. Kolom yang bertipe int64 meliputi Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, Age, dan Outcome, sedangkan kolom BMI dan DiabetesPedigreeFunction bertipe float64 karena mengandung nilai desimal. Selain itu, seluruh kolom memiliki 768 data non-null, yang berarti tidak terdapat nilai kosong (missing values) dalam dataset ini. Struktur data sepenuhnya yaitu numerik.
